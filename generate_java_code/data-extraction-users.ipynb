```python
import numpy as np
import pandas as pd
recommendations = pd.read_csv('final_data/user_recommendations_subset.csv')
```

```python
recommendations['user_id'] = recommendations['user_id'].astype('category').cat.codes + 1
recommendations['app_id'] = recommendations['app_id'].astype('category').cat.codes + 1
recommendations.to_csv('final_data/user_recommendations_subset_clean.csv', index=False)
```

```python
unique_users = recommendations['user_id'].unique()
unique_users.sort()

usernames = []
for x in range(20):
    usernames.append("Gamer" + str(x+1))

bio = ["Game-lover", "I like playing games", "I LOVE games", "I play a lot of games", "Gaming is my life",
       "Game-lover", "I like playing games", "I LOVE games", "I play a lot of games", "Gaming is my life",
       "Game-lover", "I like playing games", "I LOVE games", "I play a lot of games", "Gaming is my life",
       "Game-lover", "I like playing games", "I LOVE games", "I play a lot of games", "Gaming is my life"]

users = pd.DataFrame()
users['user_id'] = unique_users
users['display_name'] = pd.Series(usernames)
users['biography'] = pd.Series(bio)
users.head()
```

```python
# Get Gamer Data
userCodes = []
for index, row in users.iterrows():
    userCode = "new User(" 
    userCode += "\"" + row['display_name'] + "\"" + ", "
    userCode += "\"" + row['biography'] + "\"" + ", "
    userCode += "Role.GAMER" + ")"
    userCodes.append(userCode)

initUserCode = "List<User> gamers = new ArrayList<>(Arrays.asList("
for userCode in userCodes:
    initUserCode += userCode + ", "
initUserCode = initUserCode[:-2] + "));"
initUserCode
```

```python
# Get Recommendation Data
# postRepo.save(new PostGameReview("", "Bad game", profileRepo.findProfileByUserDisplayName("Gamer1"), profileRepo.findProfileByGameTitle("Team Fortress 2"), false));
# List<PostGameReview> gameReviews = new ArrayList(Arrays.asList())
import random
review_positive = ["Great game 10/10.", "RECOMMEND!!", "I lost track of time and forgot to eat.", "Not bad, not bad.", "This is my life now."]
review_negative = ["Boring game 1/10", "DO NOT RECOMMEND!!", "Waste of time", "I hate this game", "Don't play this."]

reviewList = []
for index, row in recommendations.iterrows():
    date_split = row['date'].split("-")
    year = date_split[0]
    month = date_split[1].lstrip("0")
    day = date_split[2].lstrip("0")
    randNum = random.randint(0,4)
    reviewMsg = review_positive[randNum] if (row['is_recommended']) else review_negative[randNum] 
    
    reviewItem = "new PostGameReview("
    reviewItem += "\"\"" + ", "
    reviewItem += "\"" + reviewMsg + "\"" + ", "
    reviewItem += "profileRepo.findProfileByUserId(\"" + str(row['user_id']) + "\")" + ", "
    reviewItem += "profileRepo.findProfileByGameId(\"" + str(row['app_id']) + "\")" + ", "
    reviewItem += str(row['is_recommended']).lower() + ", "
    reviewItem += "LocalDate.of(" + str(year) + "," + str(month) + "," + str(day) + ")" + ")"
    reviewList.append(reviewItem)
    
initReviewCode = "List<PostGameReview> gameReviews = new ArrayList<>(Arrays.asList("
for reviewItem in reviewList:
    initReviewCode += reviewItem + ", "
initReviewCode = initReviewCode[:-2] + "));"
initReviewCode
```
